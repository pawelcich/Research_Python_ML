```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3])
                        
```

```python
x = np.linspace(0,10,20)
y1 = x**2
y2 = x**1.5
y3 = np.sin(x)
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.5, 10.5, -5, 105])
plt.plot(x, y1, "bd-", linewidth=2, markersize=12, label = 'First')
plt.plot(x, y2, "gd-", linewidth=2, markersize=12, label = 'Second')
plt.plot(x, y3, "rs-", linewidth=2, markersize=12, label = 'Sinus')
plt.legend(loc = 'upper left')
# plt.savefig('myplot.pdf')
```

```python
x = np.linspace(0,10,20)
y1 = x**2
y2 = x**1.5
y3 = np.sin(x)
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.5, 10.5, -5, 105])
plt.loglog(x, y1, "bd-", linewidth=2, markersize=12, label = 'First')
plt.loglog(x, y2, "gd-", linewidth=2, markersize=12, label = 'Second')
plt.plot(x, y3, "rs-", linewidth=2, markersize=12, label = 'Sinus')
plt.legend(loc = 'upper left')
# plt.savefig('myplot.pdf')
```
